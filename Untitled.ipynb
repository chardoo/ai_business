```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import warnings

%matplotlib inline
raw_data = pd.read_csv('sales_data_sample.csv', encoding = 'unicode_escape')
raw_data.drop(['ORDERNUMBER' ], axis=1, inplace=True)
raw_data.drop(['ADDRESSLINE2'], axis=1, inplace=True)
raw_data['ORDERDATE'] = pd.to_datetime(raw_data['ORDERDATE'], errors='coerce')
```

```python

# Dealsize need to be encoded to numeric value using one hot encoding 

# Time-based Campaigns:  ORDERDATE to analyze the best times for promotions, such as holidays or end-of-quarter sales.
# Targeted Offers Based on Order Size:  QUANTITYORDERED and SALES to identify customers who may be interested in bulk buying or those who can be upsold.
# Price Segmentation: Tailor PRICEEACH to design price-based promotions, like discounts on high-end products or special offers for budget-conscious customers.
# Deal Size Targeting: Develop exclusive campaigns or loyalty programs for customers making Medium or Large orders.
# Product Line-Focused Campaigns: Promote top-performing product lines and offer discounts or incentives for underperforming products to balance sales.

# Normalize the data using StandardScaler
```

```python

df_onehot_size = pd.get_dummies(raw_data["DEALSIZE"], prefix="Size")
df_onehot_productLine = pd.get_dummies(raw_data["PRODUCTLINE"], prefix="ProductLine")
label_encoder = LabelEncoder()
raw_data['PRODUCTLINE_Encoded'] = label_encoder.fit_transform(raw_data['PRODUCTLINE'])

raw_data = pd.concat([raw_data, df_onehot_size, df_onehot_productLine], axis=1)
raw_data


feature_data = raw_data[['MONTH_ID', 'YEAR_ID', 'QUANTITYORDERED', 'SALES', 'PRICEEACH', 'Size_Medium', 'Size_Small',  'Size_Large', 
                        #  'PRODUCTLINE_Encoded'
                         'ProductLine_Planes', 'ProductLine_Ships', 'ProductLine_Trains',
                            'ProductLine_Trucks and Buses', 'ProductLine_Vintage Cars',
                            'ProductLine_Classic Cars',
                            'ProductLine_Motorcycles' 
                         
                         ]]
feature_data
```

```python

# Convert the scaled features into a DataFrame for easier inspection
scaler = StandardScaler()
scaled_features = scaler.fit_transform(feature_data)

scaled_data = pd.DataFrame(scaled_features, columns=feature_data.columns)
```

```python

wcss = []
k_range = range(1, 11)  # Test k from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_k = 3  
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(scaled_data)

scaled_data['Cluster'] = cluster_labels
scaled_data
```

```python
if scaled_data.shape[1] > 3:  # If more than 2 features + cluster column
    pca = PCA(n_components=2)
    pca_data = pca.fit_transform(scaled_data.drop('Cluster', axis=1))
    
    component_df = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2'],
    index=feature_data.columns
    )
    print("Principal Component Loadings:")
    print(component_df.round(3))
    print(f"\nPC1 explains {pca.explained_variance_ratio_[0]:.1%} of variance")
    print(f"PC2 explains {pca.explained_variance_ratio_[1]:.1%} of variance")
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(pca_data[:, 0], pca_data[:, 1], c=cluster_labels, cmap='viridis')
    plt.colorbar(scatter)
    plt.title('K-Means Clustering Results (PCA)')
    plt.xlabel('First Principal Component')
    plt.ylabel('Second Principal Component')
    plt.show()
    
```

```python
print("=== CLUSTER ANALYSIS ===\n")

# Check cluster sizes
unique, counts = np.unique(cluster_labels, return_counts=True)
print("Cluster Sizes:")
for i, count in enumerate(counts):
    print(f"Cluster {i}: {count} samples ({count/len(cluster_labels)*100:.1f}%)")

# Add cluster labels to your original data
clustered_data = feature_data.copy()
clustered_data['Cluster'] = cluster_labels

# Examine cluster centers (in original scale)
cluster_centers_scaled = kmeans.cluster_centers_  # Your kmeans model
cluster_centers_original = scaler.inverse_transform(cluster_centers_scaled)

centers_df = pd.DataFrame(cluster_centers_original, columns=feature_data.columns)
centers_df.index = [f'Cluster {i}' for i in range(len(centers_df))]
print("\nCluster Centers (Original Scale):")
print(centers_df.round(2))
```
